# Batch Learning vs Online Learning
- Batch Learning
    - trained using all the available data, done offline, daily or weekly
    - requires a lot of computing resources
    - can use some algorithms that are capable of learning incrementally
- Online Learning (Done offline as well, incremental learning)
    - feeding new data sequentially or by mini-batches
    - learning step is fast and cheap

# Instance-based versus Model-Based Learning
- Instance-based learning (e.g. K-NN)
    - Measure of similarity
- Model-based Learning (e.g. Regression)
    - Build a model and use it predict

- utility function (fitness function) that measures how good your model is
- cost function that measures how bad your model is.
- People typically use a cost function and try to minimize this distance (e.g. linear regression)

# Main challenges of Machine Learning
- bad data
    - Even for very simple problems, it needs thousands of examples
    - For complex problems such as image or speech recognition, need millions of examples (unless resuing existing models)
    - Sampling bias (e.g. sampling noise, nonresposne bias)
    - Poor Quality Data
        - full of errors, outliers, and noise (due to poor-quality measruemnts)
    - Irrelavent features
        - garbage in, garbage out
        - feature engineering, feature selection, feature extraction, new feature
- bad algorithm
    - overfitting the training data, not generalize well
        - simplify models by selecting one with fewer parameters, reducing number of attributes
        - gather more training data
        - reduce noise in the training data (fix data erros and remove outliers)
        - amount of regularization -> controlled by hyperparamaters
    - underfitting the training data, too simple
        - seelcting a more powerful model, with more parameters
        - feeding better features (feature engineering0
        - reducing the constaints on the model ( reducing the regularization hyperparameters)

The unreasonable effectiveness of data


Nonrepresentaive Training Data


- If Sample is too small -> sampling noise (nonrepresentative data as a result of chance)
- Even very large samples can be nonrepresentative if the sampling method is flawed. (Sampling bias)

Testing and Validating
- Split - training set and testing set (80%:20%)
- error rate one new cases : generalization error (out-of-sample error)
    - estimation of this error on testing set
- If training error is low, but generation error is high -> overfitting
- second holdout set : validation set (use for hyperparameters)
- better method: cross-validation

No Free Lunch Theorem
- There is no model that is a priori guarantted to work better. 
- The only way is to evaluate them all.
- In practice, Make reasonable assumptions, evaluate only a few.

**Select a Performance Measure**
- RMSE (Root Mean Square Error)
    - Euclidean norm (L2 distance) (higher level of norm index, more sensitve to outlier due to focus on large values and neglect small ones)    
    - When outliers are exponentially rare (Like in a bell-shaped curved), it performs very well and is generally preferred
- MAE (Mean Absolute Error or Average Absolute Deviation) if there are many outliers
    - Manhattan nrom (L1 distance)
